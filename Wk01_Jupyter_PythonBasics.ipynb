### `list` Data Type

- a list allows iteration over multiple items, facilitating automated processing 

```python
# Assign a list of integers to a variable named years

years = [2021, 2022, 2023, 2024, 2025]  

type(years)   # The data type is list
```

```python
# Like a string of characters, a list has a length too

len(years)
```

```python
# Similarly, we can slice a list to get specific elements from it

years[2:4]    # Get elements from index 2 up to (but not including) index 4

# years == [2021, 2022, 2023, 2024, 2025]
# Index:     0     1     2     3     4
```

```python
# Get the first element (index 0) from the list years
Year = years[0]   

Year    # Display the value of the variable Year
```

### Conditional Statements in Python

```python
# Basic conditional statement using if ... else ...

if Year == 2023:
    print("The value of the Year variable is 2023!")
else:
    print(f"The value of the Year variable is {Year}, not 2023!")

# Note: The f before the string in the print() function allows embedding variable values directly within the string using curly braces {}
```

```python
# If statement without else

if Year == 2023:
    print("The value of the Year variable is 2023!") 

Year   # Display the value of the variable Year again
```

```python
# If statement without else is the same as including an else part that does nothing

if Year == 2023:
    print("The value of the Year variable is 2023!")
else:
    None   # Do nothing in the else part

# Note: None is a special constant in Python that represents the absence of a value or a null value
```

---
 
### Control flow statements in Python

- For-loop lets you automate repetitive tasks!

```python
for yr in years:
    # You can do complicated operations inside the loop
    print(yr)
```

- Pairing for-loops with conditional statements allows conditional automation!

```python
for yr in years:
    if yr == 2023:
        print("Found 2023!")
    print(yr)

# The print(yr) line is outside the if block, so it runs for every year in the list
```

- Indentation is important in Python!

```python
for yr in years:
    if yr == 2023:
        print("Found 2023!")
        print(yr)

# The print(yr) line is inside the if block, so it only runs when year is 2023
```

## A Quick Overview of Variables and Data Types in Python


In Python, variables are containers used to store data values, while data types define the kind of data a variable holds.

### Python Variables

A variable is created the moment you first assign a value to it using the assignment operator (`=`).

**Dynamic Typing:** You do not need to declare a variable's type. Python automatically infers it from the value assigned.

**Reassignment:** Variables can change types during execution. For example, a variable holding an integer can later be reassigned a string.

#### Naming Rules

- Must start with a letter or an underscore (`_`)
- Cannot start with a number
- Can only contain alphanumeric characters and underscores
- Are case-sensitive (e.g., `age` and `Age` are different)
- Cannot use Python reserved keywords like `if`, `else`, or `class`

### Core Data Types

Python's built-in data types are categorized into several groups:

#### Numeric Types

- **int:** Whole numbers (e.g., `10`, `-5`)
- **float:** Decimal numbers (e.g., `3.14`)
- **complex:** Numbers with real and imaginary parts (e.g., `3 + 5j`)

#### Sequence Types

- **str:** Text data enclosed in quotes
- **list:** Ordered, mutable collections (e.g., `[1, "apple", 3.5]`)
- **tuple:** Ordered, immutable collections (e.g., `(10, 20)`)

#### Mapping Type

- **dict:** Unordered collections of key-value pairs (e.g., `{"name": "Alice"}`)

#### Set Types

- **set:** Unordered collections of unique elements

#### Boolean Type

- **bool:** Represents `True` or `False`

#### None Type

- **NoneType:** Represents the absence of a value (`None`)

### Type Conversion & Checking

**Check Type:** Use the `type()` function (e.g., `type(x)`) to find a variable's data type.

**Type Casting:** Convert between types using functions like `int()`, `float()`, or `str()`.

**Type Hinting:** For better readability in 2026, developers often use Type Hints (e.g., `age: int = 25`) to indicate intended data types, which can be checked by tools like Mypy.
